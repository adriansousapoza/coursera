### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
import sqlite3

# Create a connection to the database
conn = sqlite3.connect('spacex.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute the SQL script to create the 'spacex' table
with open('spacex.sql', 'r') as f:
    create_table_sql = f.read()
    cursor.executescript(create_table_sql)

# Execute a sample SQL query to retrieve data from the 'spacex' table
cursor.execute('SELECT * FROM spacex')
results = cursor.fetchall()

# Print the results of the SQL query
for row in results:
    print(row)

# Close the connection to the database
conn.close()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
SELECT *
FROM spacex
WHERE launch_site LIKE 'CCA%'
LIMIT 5;
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
SELECT SUM(payload_mass_kg) AS total_payload_mass_kg
FROM spacex
WHERE customer = 'NASA (CRS)' AND payload_mass_kg IS NOT NULL;
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
SELECT AVG(payload_mass_kg) AS avg_payload_mass_kg
FROM spacex
WHERE booster_version = 'F9 v1.1' AND payload_mass_kg IS NOT NULL;
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
SELECT MIN(launch_date_utc) AS first_successful_landing_date
FROM spacex
WHERE landing_outcome = 'Success (ground pad)';
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
SELECT booster_version, payload_mass_kg, landing_outcome
FROM spacex
WHERE landing_outcome LIKE 'Success (drone ship)%' 
    AND payload_mass_kg > 4000 
    AND payload_mass_kg < 6000;
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
SELECT mission_outcome, COUNT(*) AS count
FROM spacex
GROUP BY mission_outcome;
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
SELECT booster_version, payload_mass_kg
FROM spacex
WHERE payload_mass_kg = (SELECT MAX(payload_mass_kg) FROM spacex)
```

### Task 9


##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
SELECT landing_outcome, booster_version, launch_site
FROM spacex
WHERE YEAR(launch_date_utc) = 2015 AND landing_outcome LIKE 'Failure (drone ship)%';
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
SELECT landing_outcome, COUNT(*) AS count
FROM spacex
WHERE launch_date_utc BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY landing_outcome
ORDER BY count DESC;
```
